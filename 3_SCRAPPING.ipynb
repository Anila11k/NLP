```python
from bs4 import BeautifulSoup
import requests 
url="https://www.geeksforgeeks.org/geek-week-2023-score-a-career-six-with-gfg-courses/"
response= requests.get(url)
soup=BeautifulSoup(response.text,"html.parser")
print(soup.title.text)
```

```python
paragraphs = [p.text.strip() for p in soup.find_all("p")]
paragraphs
```

```python
headings = [h2.text.strip() for h2 in soup.find_all("h2")]
```

```python
headings
```

```python
headings= soup.find_all(["h1","h2","h3"])
for heading in headings:
    print (heading.text.strip())
```

```python
links = soup.find_all("a")
for link in links:
    href = link.get("href")
    if href:
        print(href)
```

```python
tables = soup.find_all("tables")
for table in tables:
    print(table.text.strips())
```

```python
main_content = soup.find("div",{"id": "main"})
print(main_content.text.strip()) if main_content else print("no main contents found")
```

```python
lists = soup.find_all(["ul", "ol"])
for lst in lists:
    items = lst.find_all("li")
    for item in items:
        print(item.text.strip())
```

```python
images = soup.find_all("img")
for  img in images:
    src = img.get("src")
    if src:
        print(src)
```
