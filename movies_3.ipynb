```python
import pandas as pd
import numpy as np
df=pd.read_csv('Top_rated_movies1.csv')
df.head()
```

```python
df.tail
```

```python
Top_movie=df
Top_movie
```

```python
Top_movie.isna().sum()
```

```python
Top_movie.dropna(inplace=True)
```

```python
Top_movie.isna().sum()
```

```python
c=Top_movie['vote_average'].mean()
c
```

```python
m=Top_movie['vote_count'].quantile(0.9)
m
```

```python
movies=Top_movie.copy().loc[Top_movie['vote_count']>=m]
movies.shape
```

```python
def weighted_rating(x,m=m,c=c):
    v=x['vote_count']
    r=x['vote_average']
    return (v/(v+m)*r)+(m/(v+m)*c)
```

```python
movies['score'] = movies.apply(weighted_rating,axis=1)
movies.head(10)
```

```python
movies[['title','vote_count','vote_average','score']].head(10)
```

```python
movies[['overview']]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')
movies['overview']=movies['overview'].fillna(' ')
```

```python
tfidf_matrix=tfidf.fit_transform(movies['overview'])
print(tfidf_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
movie_similarity=cosine_similarity(tfidf_matrix)
movie_similarity
```
